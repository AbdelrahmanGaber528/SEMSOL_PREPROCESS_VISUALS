```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.preprocessing import StandardScaler
```

```python
X_train_resampled = joblib.load('data/X_train_custom_resampled.joblib')
y_train_resampled = joblib.load('data/y_train_custom_resampled.joblib')
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_resampled)
```

```python
joblib.dump(X_train_scaled, 'data/X_train_scaled.joblib')
joblib.dump(scaler, 'data/scaler.joblib')
```

```python
df_scaled = pd.DataFrame(X_train_scaled, columns=X_train_resampled.columns)
corr_matrix = df_scaled.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Scaled Features')
plt.show()
```

```python
df_scaled['label'] = y_train_resampled

for feature in X_train_resampled.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='label', y=feature, data=df_scaled)
    plt.title(f'Distribution of {feature} by Engagement Level')
    plt.show()
```
